```python
import speech_recognition as sr
import pyttsx3
import datetime
import webbrowser
import random
import os
import subprocess
import requests

# Initialize TTS engine and recognizer
engine = pyttsx3.init()
recognizer = sr.Recognizer()

# Your OpenWeatherMap API key
WEATHER_API_KEY = 'd7e0043c292d573bc55a2d511d8161f2'#ur api key

def speak(text):
    """Make the assistant speak."""
    engine.say(text)
    engine.runAndWait()

def ai_command():
    """Listen for and recognize speech commands."""
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
        try:
            query = recognizer.recognize_google(audio)
            print(f"Command: {query}")
            return query.lower()
        except sr.UnknownValueError:
            speak("Sorry, I didn't understand that.")
            return ""
        except sr.RequestError:
            speak("Sorry, there was an issue with the speech recognition service.")
            return ""

def greet_user():
    """Greet the user based on the time of day."""
    hour = datetime.datetime.now().hour
    if hour < 12:
        speak("Good morning!")
    elif hour < 18:
        speak("Good afternoon!")
    else:
        speak("Good evening!")
    speak("How can I assist you today?")

def tell_joke():
    """Tell a random joke."""
    jokes = [
        "Why don't scientists trust atoms? Because they make up everything!",
        "Why did the computer go to the doctor? Because it had a virus!",
        "Why do programmers prefer dark mode? Because light attracts bugs!",
        "What did one wall say to the other wall? I'll meet you at the corner!"
    ]
    return random.choice(jokes)

def search_google(query):
    """Search a query on Google."""
    webbrowser.open(f"https://www.google.com/search?q={query}")
    speak(f"Here are the search results for {query}")

def open_website(site_name):
    """Open specific websites."""
    sites = {
        "youtube": "https://youtube.com",
        "facebook": "https://facebook.com",
        "twitter": "https://twitter.com",
        "instagram": "https://instagram.com",
        "linkedin": "https://linkedin.com"
    }
    url = sites.get(site_name)
    if url:
        webbrowser.open(url)
        speak(f"Opening {site_name}")
    else:
        speak("Sorry, I don't have that website in my list.")

def get_weather():
    """Fetch weather information for a specified city."""
    speak("Please tell me the city name.")
    city = ai_command()
    if city:
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={WEATHER_API_KEY}&units=metric"
        try:
            response = requests.get(url)
            data = response.json()
            if data.get('cod') == 200:
                temp = data['main']['temp']
                description = data['weather'][0]['description']
                speak(f"The current temperature in {city} is {temp} degrees Celsius with {description}.")
            else:
                speak("I couldn't find that city.")
        except:
            speak("Sorry, I couldn't fetch the weather data right now.")
    else:
        speak("City name not provided.")

def play_music():
    """Play a random music file from the directory."""
    music_directory = r"C:\Users\afnan\Music"  # Update as needed
    songs = [f for f in os.listdir(music_directory) if f.endswith(".mp3")]
    if songs:
        song = random.choice(songs)
        song_path = os.path.join(music_directory, song)
        subprocess.run(["start", "", song_path], shell=True)
        speak(f"Playing {song}")
    else:
        speak("No music files found.")

def calculate(expression):
    """Calculate a simple mathematical expression."""
    try:
        result = eval(expression)
        speak(f"The result is {result}")
    except:
        speak("Sorry, I couldn't understand the calculation.")

# Main loop
greet_user()
while True:
    command = ai_command()

    if not command:
        continue

    if 'time' in command:
        current_time = datetime.datetime.now().strftime("%H:%M")
        speak(f"Current time is {current_time}")

    elif 'open youtube' in command:
        webbrowser.open("https://youtube.com")
        speak("Opening YouTube")

    elif 'joke' in command:
        joke = tell_joke()
        speak(joke)

    elif 'search' in command:
        query = command.replace("search", "").strip()
        if query:
            search_google(query)
        else:
            speak("Please tell me what to search for.")

    elif 'play music' in command:
        play_music()

    elif 'weather' in command:
        get_weather()

    elif 'open facebook' in command:
        open_website('facebook')

    elif 'open twitter' in command:
        open_website('twitter')

    elif 'open instagram' in command:
        open_website('instagram')

    elif 'calculate' in command:
        expression = command.replace("calculate", "").strip()
        if expression:
            calculate(expression)
        else:
            speak("Please tell me the calculation to perform.")

    elif 'exit' in command or 'stop' in command or 'quit' in command:
        speak("Goodbye!")
        break

    elif 'shutdown' in command:
        speak("Shutting down. Have a nice day!")
        os.system("shutdown /s /t 1")
        break

    else:
        speak("Sorry, I can't perform that task.")
```
